```python
for i in range(9,14):
    file = f'./data/CIKM22Competition/{i}/train.pt'
    data = torch.load(file)
    file2 = f'./data/CIKM22Competition/{i}/val.pt'
    data2 = torch.load(file2)
    targets = torch.Tensor([])
    for sample in data:
        y = sample.y
        targets = torch.concat((targets,y),dim=0)
    for sample in data2:
        y = sample.y
        targets = torch.concat((targets,y),dim=0)
    std,mean = torch.std_mean(targets,dim=0)
    print(i,std,mean)
```
